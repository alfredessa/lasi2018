```python
sns.distplot(moby_len_lines, color='y',hist=False)
sns.distplot(hamlet_len_lines, color='g',hist=False)
```

```python
def word_dist(url):
    f = open(url,'r')
    lines = f.readlines()
    len_lines = [len(line) for line in lines]
    sns.distplot(len_lines,hist=False)
```

```python
url1 = "../data/mdick.txt"
url2 = "../data/hamlet.txt"
url3 = "../data/emma.txt"
```

```python
word_dist(url1)
word_dist(url2)
word_dist(url3)
```
